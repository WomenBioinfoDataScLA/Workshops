### **Cmo es el proceso de carga de datos?**
A continuacin, se detalla una breve descripcin del proceso de carga de la base de datos y de la anonimizacin de las resoluciones, con el objetivo de comprender mejor el proceso realizado por el Juzgado 10:
 


1.   Una vez que la resolucin est firmada por el Juez a cargo, la persona responsable de la mesa de entradas le asigna un nmero de registro interno a la resolucin.
2.   Luego, y previo a su publicacin, debe cumplir con el proceso de **anonimizacin**. 
El Juzgado 10 dise un proceso interno de eliminacin de los datos que pudieran identificar a las personas involucradas en los conflictos de naturaleza penal, contravencional o de faltas, o para evitar la frustracin de una investigacin judicial que se encuentre en curso. Este proceso lo realizan de forma manual.

3. Seguidamente, la persona a cargo de la anonimizacin guarda una versin de este archivo en un formato abierto de texto (.odt), en una carpeta del Drive de Google de la cuenta privada del Juzgado 10, cambiando el nombre del archivo por el nmero de registro, la fecha, y el artculo infringido. Asimismo, carga en el set de datos unificado del Juzgado 10 toda la fila de datos.
En este punto es importante remarcar que el Juzgado 10 tiene un set de datos privado -set de datos unificado del Juzgado 10- y uno pblico -set de datos con perspectiva de gnero-. A travs de una frmula de google sheet, el set de datos con perspectiva de gnero se actualiza automticamente una vez que las personas que trabajan en el Juzgado 10 cargan los datos en el set de datos privado.

4.  Como ltimo paso, la persona encargada de la comunicacin en la red social del Juzgado 10 o community manager, revisa que el proceso de anonimizacin se haya realizado correctamente. Si este proceso se cumpli correctamente, se publica la resolucin en un posteo de Twitter, la carga en el set de datos y la sube a la carpeta Resoluciones del Juzgado Resoluciones 2020 del Repositorio de Datos Abiertos Pblico del Juzgado 10

### **Diccionario de variables**
La base cuenta con 64 campos, que poseen informacin sobre distintas dimensiones de las cuestiones registradas por el Juzgado 10. Agrupamos estas variables en cuatro ejes que sintetizamos a continuacin. Para realizar el anlisis de la base tomamos algunas variables (no su totalidad). A continuacin detallaremos solamente aquellas utilizadas en este informe:

**Datos sobre las Resoluciones:**

Este conjunto de variables presenta informacin sobre las resoluciones propiamente dichas. 
Los campos que utilizamos para el anlisis son:

```Materia```: Los valores que toma esta variable pueden ser penal, contravencional, faltas, amparo, habeas corpus o ejecuciones de multa.

```Artculo infringido```: Los artculos infringidos en el caso

```Cdigo o ley```: indica si el artculo pertenece al Cdigo Penal de la Nacin, Cdigo Contravencional, Ley 451 - Rgimen de Faltas de la Ciudad de Buenos Aires, Ley 23737 - Tenencia y trfico de estupefacientes, Ley 13944 - Incumplimiento de los deberes de asistencia familiar, Ley 24720 - Impedimento de contacto con padre no conviviente, Ley 14346 - Malos tratos o actos de crueldad a los animales, Ley 26735 - Rgimen Penal Tributario, Ley 12331 - Ley Nacional de Profilaxis.

```Conducta```: Para casos de violencia de gnero aparecen conductas vinculadas amenazas, hostigamiento, lesiones, acoso sexual callejero, daos, abuso sexual, ingreso a domicilio sin autorizacin, entre otros.

```Conducta descripcion```: se especifica si el delito, la contravencin o la falta infringida tiene alguna particularidad, como puede ser que se encuentra agravada por alguna causal

**Datos sobre las partes involucradas (persona agresora y persona vctima) y la relacin entre ellas:**

El segundo eje presenta diferentes caractersticas de la persona acusada y de la denunciante: su edad, gnero, nacionalidad y nivel de instruccin. Asimismo explicita la relacin entre estas personas y si tienen hijos/as en comn. Las variables utilizadas son:

```Gnero acusado/a```: toma los valores binarios de varn o mujer (para casos de violencia de gnero el valor es varn) segn el registro de la Oficina de Violencia Domstica, la recoleccin de este dato no depende del Juzgado 10.

```Nacionalidad acusado/a```: indica cul es la nacionalidad de la persona acusada (argentina, boliviana, paraguaya, chilena, dominicana, etc.)
Edad acusado/a al momento del hecho: valor numrico (toma valores de nmeros enteros)

```Nivel de instruccin acusado/a```: est compuesto por diferentes categoras fijas que van desde primario incompleto a universitario completo.

```Gnero denunciante```: toma los valores binarios de varn o mujer (para casos de violencia de gnero el valor es mujer) segn el registro de la Oficina de Violencia Domstica, la recoleccin de este dato no depende del Juzgado 10.

```Nacionalidad denunciante```: indica cul es la nacionalidad de la persona acusada (argentina, boliviana, paraguaya, chilena, dominicana, etc.).

```Edad denunciante al momento del hecho```: valor numrico (toma valores de nmeros enteros)

```Nivel de instruccin denunciante```: est compuesto por diferentes categoras fijas que van desde primario incompleto a universitario completo.

```Domicilio denunciante```: indica la zona de CABA -norte, sur, este, oeste- de la persona que denuncia.

```Relacin y tipo entre acusado/a y denunciante```: indica el tipo de vnculo que tiene la persona acusada con la denunciante, en general para casos de violencia de gnero es ex pareja, pareja, vecino, familiar, entre otros.

```Hijos hijas en comn```: valor booleano si o no si corresponde

**Datos de Violencia de gnero:** 

Indican la presencia o no en el caso del contexto de violencia de gnero y detallan los distintos tipos que se encuentran presentes, incluyendo su modalidad, lugar donde ocurri y frecuencia, as como las frases de la agresin verbal sufridas por la denunciante.

```Violencia de gnero```: si el hecho objeto de investigacin se encuentra dentro de un contexto de violencia de gnero o no. Toma los valores s o no. Es importante aclarar que los distintos tipos de violencia que denota un caso, no se desprenden necesaria y exclusivamente de la conducta y/o artculo infringido, sino del contexto en el que tuvo lugar, segn el contenido del informe de la OVD o de la denuncia que da inicio al caso.

```Violencia fsica```: indica si por la declaracin de la vctima hubo violencia de esa ndole. En el set de datos las opciones son si, no, s/d (sin datos), y no_corresponde. La diferencia entre las variables no y no_corresponde, es que la primera de ellas indica que en un caso con contexto de violencia de gnero no hubo violencia de esa ndole, y la otra se utiliza cuando en el caso no hay un contexto de violencia de gnero.

```Violencia psicolgica```: idem anterior

```Violencia econmica```: idem anterior

```Violencia sexual```: idem anterior

```Violencia social```: idem anterior

```Modalidad de la violencia```: es la forma en que se manifiestan los distintos tipos de violencia contra las mujeres en diferentes mbitos. ```Frecuencia episodios```: se indica carcter espordico (cuando las agresiones ocurren de forma aislada), diario, habitual (si ocurren semanalmente), eventual (si se dan quincenal o mensualmente), o si se trata de la primera agresin sufrida.

```Zona del hecho```: Dividido en zonas geogrficas de la CABA como Norte (compuesta por Comunas 12, 13, 14 y 15), Sur (compuesta por Comunas 4, 8 y 9), Este (compuesta por Comunas 1, 2 y 3) y Oeste (compuesta por Comunas 5, 6, 7, 10 y 11)

```Lugar del hecho```: lugar fsico (va pblica, domicilio particular, medios tecnolgicos)

```Medidas de proteccin vigentes al momento del hecho```: indica las medidas de proteccin que se hayan impuesto para proteger a la vctima y si estaban vigentes al momento de los hechos. Toma valores de si o no
