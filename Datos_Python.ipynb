## **El Juzgado 10**

El trabajo realizado por el [Juzgado Penal, Contravencional y de Faltas N° 10](https://twitter.com/jpcyf10) de la Ciudad Autónoma de Buenos Aires (a partir de ahora Juzgado 10) nos ayuda a comprender mejor los tipos, formas, modalidades que tiene la violencia de género, en la medida en que disponibiliza su información en formatos abiertos y accesibles.

Este Juzgado posee una base de datos abierta y con perspectiva de género que tiene el objetivo de disponibilizar, a través de medios digitales, todas las resoluciones y sentencias del tribunal. El [Repositorio de Datos Abiertos](https://drive.google.com/drive/folders/0B9wNhp3GjjazZ2VCQVZmM3MwTTQ?resourcekey=0-s10MrFRR0gmOd1TgrDZs7g)
con el que cuenta, contiene entre otros documentos, el Set de Datos Abiertos
 donde están disponibles todas las resoluciones desde agosto del 2016 hasta la actualidad, y detallan minuciosamente cada una de ellas para que sea más fácil y práctico poder analizarlas. El 3 de junio de 2020, en el contexto del “Ni Una Menos”, publicaron su nueva base de datos unificada con perspectiva de género


### **¿Cómo es el proceso de carga de datos?**
A continuación, se detalla una breve descripción del proceso de carga de la base de datos y de la anonimización de las resoluciones, con el objetivo de comprender mejor el proceso realizado por el Juzgado 10:
 


1.   Una vez que la resolución está firmada por el Juez a cargo, la persona responsable de la mesa de entradas le asigna un número de registro interno a la resolución.
2.   Luego, y previo a su publicación, debe cumplir con el proceso de **anonimización**. 
El Juzgado 10 diseñó un proceso interno de eliminación de los datos que pudieran identificar a las personas involucradas en los conflictos de naturaleza penal, contravencional o de faltas, o para evitar la frustración de una investigación judicial que se encuentre en curso. Este proceso lo realizan de forma manual.

3. Seguidamente, la persona a cargo de la anonimización guarda una versión de este archivo en un formato abierto de texto (.odt), en una carpeta del Drive de Google de la cuenta privada del Juzgado 10, cambiando el nombre del archivo por el número de registro, la fecha, y el artículo infringido. Asimismo, carga en el set de datos unificado del Juzgado 10 toda la fila de datos.
En este punto es importante remarcar que el Juzgado 10 tiene un set de datos privado -set de datos unificado del Juzgado 10- y uno público -set de datos con perspectiva de género-. A través de una fórmula de google sheet, el set de datos con perspectiva de género se actualiza automáticamente una vez que las personas que trabajan en el Juzgado 10 cargan los datos en el set de datos privado.

4.  Como último paso, la persona encargada de la comunicación en la red social del Juzgado 10 o community manager, revisa que el proceso de anonimización se haya realizado correctamente. Si este proceso se cumplió correctamente, se publica la resolución en un posteo de Twitter, la carga en el set de datos y la sube a la carpeta “Resoluciones del Juzgado” “Resoluciones 2020” del Repositorio de Datos Abiertos Público del Juzgado 10

### **Diccionario de variables**
La base cuenta con 64 campos, que poseen información sobre distintas dimensiones de las cuestiones registradas por el Juzgado 10. Agrupamos estas variables en cuatro ejes que sintetizamos a continuación. Para realizar el análisis de la base tomamos algunas variables (no su totalidad). A continuación detallaremos solamente aquellas utilizadas en este informe:

**Datos sobre las Resoluciones:**

Este conjunto de variables presenta información sobre las resoluciones propiamente dichas. 
Los campos que utilizamos para el análisis son:

```Materia```: Los valores que toma esta variable pueden ser penal, contravencional, faltas, amparo, habeas corpus o ejecuciones de multa.

```Artículo infringido```: Los artículos infringidos en el caso

```Código o ley```: indica si el artículo pertenece al Código Penal de la Nación, Código Contravencional, Ley 451 - Régimen de Faltas de la Ciudad de Buenos Aires, Ley 23737 - Tenencia y tráfico de estupefacientes, Ley 13944 - Incumplimiento de los deberes de asistencia familiar, Ley 24720 - Impedimento de contacto con padre no conviviente, Ley 14346 - Malos tratos o actos de crueldad a los animales, Ley 26735 - Régimen Penal Tributario, Ley 12331 - Ley Nacional de Profilaxis.

```Conducta```: Para casos de violencia de género aparecen conductas vinculadas amenazas, hostigamiento, lesiones, acoso sexual callejero, daños, abuso sexual, ingreso a domicilio sin autorización, entre otros.

```Conducta descripcion```: se especifica si el delito, la contravención o la falta infringida tiene alguna particularidad, como puede ser que se encuentra agravada por alguna causal

**Datos sobre las partes involucradas (persona agresora y persona víctima) y la relación entre ellas:**

El segundo eje presenta diferentes características de la persona acusada y de la denunciante: su edad, género, nacionalidad y nivel de instrucción. Asimismo explicita la relación entre estas personas y si tienen hijos/as en común. Las variables utilizadas son:

```Género acusado/a```: toma los valores binarios de varón o mujer (para casos de violencia de género el valor es varón) según el registro de la Oficina de Violencia Doméstica, la recolección de este dato no depende del Juzgado 10.

```Nacionalidad acusado/a```: indica cuál es la nacionalidad de la persona acusada (argentina, boliviana, paraguaya, chilena, dominicana, etc.)
Edad acusado/a al momento del hecho: valor numérico (toma valores de números enteros)

```Nivel de instrucción acusado/a```: está compuesto por diferentes categorías fijas que van desde primario incompleto a universitario completo.

```Género denunciante```: toma los valores binarios de varón o mujer (para casos de violencia de género el valor es mujer) según el registro de la Oficina de Violencia Doméstica, la recolección de este dato no depende del Juzgado 10.

```Nacionalidad denunciante```: indica cuál es la nacionalidad de la persona acusada (argentina, boliviana, paraguaya, chilena, dominicana, etc.).

```Edad denunciante al momento del hecho```: valor numérico (toma valores de números enteros)

```Nivel de instrucción denunciante```: está compuesto por diferentes categorías fijas que van desde primario incompleto a universitario completo.

```Domicilio denunciante```: indica la zona de CABA -norte, sur, este, oeste- de la persona que denuncia.

```Relación y tipo entre acusado/a y denunciante```: indica el tipo de vínculo que tiene la persona acusada con la denunciante, en general para casos de violencia de género es “ex pareja”, “pareja”, “vecino”, “familiar”, entre otros.

```Hijos hijas en común```: valor booleano “si” o “no” si corresponde

**Datos de Violencia de género:** 

Indican la presencia o no en el caso del contexto de violencia de género y detallan los distintos tipos que se encuentran presentes, incluyendo su modalidad, lugar donde ocurrió y frecuencia, así como las frases de la agresión verbal sufridas por la denunciante.

```Violencia de género```: si el hecho objeto de investigación se encuentra dentro de un contexto de violencia de género o no. Toma los valores “sí” o “no”. Es importante aclarar que los distintos tipos de violencia que denota un caso, no se desprenden necesaria y exclusivamente de la conducta y/o artículo infringido, sino del contexto en el que tuvo lugar, según el contenido del informe de la OVD o de la denuncia que da inicio al caso.

```Violencia física```: indica si por la declaración de la víctima hubo violencia de esa índole. En el set de datos las opciones son “si”, “no”, “s/d” (sin datos), y “no_corresponde”. La diferencia entre las variables “no” y “no_corresponde”, es que la primera de ellas indica que en un caso con contexto de violencia de género no hubo violencia de esa índole, y la otra se utiliza cuando en el caso no hay un contexto de violencia de género.

```Violencia psicológica```: idem anterior

```Violencia económica```: idem anterior

```Violencia sexual```: idem anterior

```Violencia social```: idem anterior

```Violencia ambiental```: idem anterior

```Violencia simbólica```: idem anterior

```Frases agresión```: variable de texto libre donde se encuentra la transcripción de las frases descriptas por la víctima como la agresión verbal sufrida y que son parte de los hechos del caso. Aplica para los casos de violencia verbal.

```Modalidad de la violencia```: es la forma en que se manifiestan los distintos tipos de violencia contra las mujeres en diferentes ámbitos. Las opciones dentro del set son: doméstica, institucional, mediática, laboral, contra la libertad reproductiva, obstétrica, en espacio público o privado, y política y pública.

```Frecuencia episodios```: se indica carácter esporádico (cuando las agresiones ocurren de forma aislada), diario, habitual (si ocurren semanalmente), eventual (si se dan quincenal o mensualmente), o si se trata de la primera agresión sufrida.

```Zona del hecho```: Dividido en zonas geográficas de la CABA como Norte (compuesta por Comunas 12, 13, 14 y 15), Sur (compuesta por Comunas 4, 8 y 9), Este (compuesta por Comunas 1, 2 y 3) y Oeste (compuesta por Comunas 5, 6, 7, 10 y 11)

```Lugar del hecho```: lugar físico (vía pública, domicilio particular, medios tecnológicos)

```Medidas de protección vigentes al momento del hecho```: indica las medidas de protección que se hayan impuesto para proteger a la víctima y si estaban vigentes al momento de los hechos. Toma valores de “si” o “no”

**Datos sobre Tribunales Superiores** (Cámara de Apelaciones PCyF, Tribunal Superior de Justicia CABA, Corte Suprema de Justicia de la Nación): 



Agrupa los datos referidos a aspectos procesales, incluyendo qué decisiones se tomaron en otras instancias, números de registros relacionados a estas resoluciones, y links de acceso a las mismas. Ello, siempre y cuando la resolución del Juzgado 10 haya sido recurrida. Estas variables no fueron utilizadas en nuestro análisis, si se desea consultarlas se encuentran en el Diccionario de datos presentado por el Juzgado 10 en su repositorio.

En este recorrido vamos a adentrarnos en el mundo de los datos, y para ello utilizaremos Pandas, una biblioteca de Python que nos permite trabajar con archivos de formato definido: CSV, un excel, etc. Además, Pandas proporciona estructuras de datos rápidas, flexibles y expresivas diseñadas para que trabajar con datos "relacionales" o "etiquetados" sea fácil e intuitivo. Pandas soporta múltipes tipos de datos:

- Datos tabulares con columnas de tipo heterogéneo, como en una tabla SQL o en una hoja de cálculo de Excel.
- Datos ordenados y desordenados (no necesariamente frecuencia fija).
- Datos matriciales arbitrarios (homogéneamente tipados o heterogéneos) con etiquetas de fila y columna.
- Cualquier otra forma de conjuntos de datos observacionales/estadísticos. 

Los datos en realidad no necesitan ser etiquetados para ser colocados en una estructura de datos de pandas. Estas estructuras se construyen a partir de arrays (listas), pero agregando nuevas funcionalidades. Pandas maneja dos estructuras de datos: Series y DataFrames.

### **Series (1-dimensional)**


Las series pueden contener cualquier tipo de datos (enteros, cadenas, números de punto flotante, etc.). Y se pueden crear del siguiente modo:


````python
import pandas as pd
una_serie = pd.Series(['Peru', 'Argentina', 'Bolivia', 'Uruguay', 'Brasil', 'Chile'], dtype='string')

print(una_serie)
````

### **DataFrame (2-dimensiones)**

Un **DataFrame** es una estructura tabular bidimensional de datos tabulares, potencialmente heterogéneos, con ejes etiquetados (filas y columnas). 

Podemos crear un DataFrame del sigueinte modo:

````python
paises_latam = pd.DataFrame(data ={"Pais": ['Peru', 'Argentina'], "Lengua oficial": ['Español', 'Español']}, index = [1,2])

print(paises_latam)
````
